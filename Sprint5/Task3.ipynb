Создать базовый класс Review в C#, который будет представлять отзывы о
продуктах или услугах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Review:

• Атрибуты: ID отзыва (ReviewId), Текст отзыва (Text), Рейтинг (Rating).

• Методы:

        • DisplayReview(): метод для отображения отзыва.

        • RateProduct(): метод для присвоения рейтинга продукту.

        • GetReviewDetails(): метод для получения деталей отзыва.

Требования к производным классам:
1. ОтзывОбслуживания (ServiceReview): Должен содержать дополнительные
атрибуты, такие как Дата посещения (VisitDate).
Метод DisplayReview() должен быть переопределен для включения даты
посещения в отображение отзыва.

2. ОтзывТовара (ProductReview): Должен содержать дополнительные атрибуты,
такие как Идентификатор продукта (ProductId). Метод RateProduct() должен
быть переопределен для связывания рейтинга с конкретным продуктом.

3. ОтзывУслуги (ServiceReview) (если требуется третий класс): Должен
содержать дополнительные атрибуты, такие как Время начала услуги
(StartTime). Метод GetReviewDetails() должен быть переопределен для
отображения времени начала услуги вместе с другими деталями отзыва.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) используйтие в проекте коллекции, делегаты, события.

```
using System;
using System.Collections.Generic;
using System.Reflection;
using System.Linq;

//ДОБАВЛЕНИЕ ДЕЛЕГАТА
public delegate void DisplayDelegate();

public abstract class Review //базовый класс
{
    protected int _reviewId { get; set; } // ID отзыва 
    public int ReviewId
    {
        get { return _reviewId; }
        set
        {
            if (value >= 0)
                _reviewId = value;
            else
                throw new ArgumentOutOfRangeException("ID отзыва не может быть отрицательным!");
        }
    }

    protected string _text { get; set; }  // Текст отзыва
    public string Text
    {
        get { return _text; }
        set { _text = value; }
    }

    protected double _rating { get; set; }  // Рейтинг
    public double Rating
    {
        get { return _rating; }
        set
        {
            if (value >= 0)
                _rating = value;
            else
                throw new ArgumentOutOfRangeException("Рейтинг отзыва не может быть отрицательным!");
        }
    }

    protected string _author { get; set; } // Автор отзыва
    public string Author
    {
        get { return _author; }
        set { _author = value; }
    }

    protected DateTime _dateCreated { get; set; } // Дата создания отзыва
    public DateTime DateCreated
    {
        get { return _dateCreated; }
        set { _dateCreated = value; }
    }

    protected string _createdDate { get; set;} // доп атрибут для дальнейшей перегрузки метода
    public string CreatedDate
    {
        get { return _createdDate; }
        set { _createdDate = value;}
    }

    protected string _productName { get; set; } // Название продукта/услуги
    public string ProductName
    {
        get { return _productName; }
        set { _productName = value; }
    }

    protected string _tags { get; set; } // новый атрибут - тэги отзыва
    public string Tags
    {
        get { return _tags; }
        set { _tags = value; }
    }

    public Review(int _ReviewId, string _Text, double _Rating, string _Author, DateTime _DateCreated, string _ProductName, string _Tags)
    {
        ReviewId = _ReviewId;
        Text = _Text;
        Rating = _Rating;
        Author = _Author;
        DateCreated = _DateCreated;
        ProductName = _ProductName;
        Tags = _Tags;
    }

    public Review(int _ReviewId, string _Text, double _Rating, string _Author, string _CreatedDate, string _ProductName, string _Tags) // перегрузка метода
    {
        ReviewId = _ReviewId;
        Text = _Text;
        Rating = _Rating;
        Author = _Author;
        CreatedDate = _CreatedDate;
        ProductName = _ProductName;
        Tags = _Tags;
    }

    public virtual string DisplayReview() // метод для отображения отзыва
    {
        return $"ID: {ReviewId}, Текст отзыва: {Text}, Автор: {Author}, Дата: {DateCreated.ToString("dd.MM.yyyy")}";
    }

    public void RateProduct(double _Rating) //метод для присвоения рейтинга продукту
    {
        Rating = _Rating;
    }

    public virtual string GetReviewDetails()  //метод для получения деталей отзыва.
    {
        return ($"Отзыв: {Text}, Тэги отзыва: {Tags}, Рейтинг: {Rating}, Автор: {Author}, Дата: {DateCreated.ToString("dd.MM.yyyy")}, Продукт/Услуга: {ProductName}.");
    }

    public virtual string GetReviewDetail()  //метод для получения деталей отзыва.
    {
        return ($"Отзыв: {Text}, Тэги отзыва: {Tags}, Рейтинг: {Rating}, Автор: {Author}, Дата: {CreatedDate}, Продукт/Услуга: {ProductName}.");
    }

    //protected static void DisplayMethodCount(object obj)
    //{
    //    Type type = obj.GetType();
    //    MethodInfo[] methods = type.GetMethods(BindingFlags.Public | BindingFlags.Instance | BindingFlags.DeclaredOnly);
    //    Console.WriteLine($"Количество доступных методов для этого экземпляра класса {type.Name}: {methods.Length}");
    //}

    //ДОБАВЛЕНИЕ КОЛЛЕКЦИИ HASHSET
    public override bool Equals(object obj)
    {
        // Простая проверка равенства базируется на авторе, рейтинге отзыва и продукта, к которому он оставлен
        if (obj is Review review)
        {
            return Author == review.Author && Rating == review.Rating && ProductName == review.ProductName;
        }
        return false;
    }
    public override int GetHashCode()
    {
        // Создание хеш-кода на основе автора и рейтинга
        return HashCode.Combine(Author, Rating, ProductName);
    }
    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Автор отзыва: {Author}, Рейтинг отзыва: {Rating}, Продукт/услуга, к которому оставлен отзыв: {ProductName}");
    }
    public virtual void DisplayTag()
    {
        Console.WriteLine($"Автор отзыва: {Author}, Рейтинг отзыва: {Rating}, Тэг отзыва: {Tags}");
    }

}

public interface IServiceCare
{
    void DisplayingReview();
    void RespondToReview(string response); // новый метод для ответа на отзыв
}

public class ServiceCare : Review, IServiceCare //Отзыв Обслуживания, множественное наследование
{
    protected string _visitDate; // дата посещения
    public string VisitDate
    {
        get { return _visitDate; }
        set { _visitDate = value; }
    }

    protected string _serviceProvider { get; set; } // Поставщик услуги
    public string ServiceProvider
    {
        get { return _serviceProvider; }
        set { _serviceProvider = value; }
    }

    protected string _serviceDescription { get; set; } // Описание услуги
    public string ServiceDescription
    {
        get { return _serviceDescription; }
        set { _serviceDescription = value; }
    }

    public ServiceCare(int _ReviewId, string _Text, double _Rating, string _Author, string _CreatedDate, string _ProductName, string _VisitDate, string _ServiceProvider, string _ServiceDescription, string _Tags)
        : base(_ReviewId, _Text, _Rating, _Author, _CreatedDate, _ProductName, _Tags)
    {
        VisitDate = _VisitDate;
        ServiceProvider = _ServiceProvider;
        ServiceDescription = _ServiceDescription;
        //DisplayMethodCount(this);
    }

    public virtual string GetServiceDetails() // метод для получения деталей об услуге
    {
        return $"Услуга: {ServiceDescription}, Поставщик: {ServiceProvider}";
    }

    public virtual void DisplayingReview() // неявная реализация интерфейса
    {
        Console.WriteLine($"{base.DisplayReview()} \nКлиент, оставивший отзыв, посетил услугу {VisitDate}.");
        // вывести количество доступных для этого экземпляра методов РЕФЛЕКСИЯ В С#
    }

    void IServiceCare.RespondToReview(string response) // явная реализация интерфейса
    {
        Console.WriteLine($"Ответ на отзыв от {Author}: {response}");
    }
}

public class ServiceRatingCare : ServiceCare //Отзыв обслуживания + присвоенный рейтинг, сложное наследование 
{
    private string _nameOfClient;
    public string NameOfClient
    {
        get { return _nameOfClient; }
        set { _nameOfClient = value; }
    }

    public ServiceRatingCare(int _ReviewId, string _Text, double _Rating, string _Author, string _CreatedDate, string _ProductName, string _VisitDate, string _ServiceProvider, string _ServiceDescription, string _NameOfClient, string _Tags)
        : base(_ReviewId, _Text, _Rating, _Author, _CreatedDate, _ProductName, _VisitDate, _ServiceProvider, _ServiceDescription, _Tags)
    {
        NameOfClient = _NameOfClient;
    }

    public override void DisplayingReview() //переопределен для отображения имени клиента
    {
        Console.WriteLine($"{NameOfClient}, оставивший отзыв, посетил услугу {VisitDate}");
    }

    public void DisplayClientName() // новый метод для отображения имени клиента
    {
        Console.WriteLine($"Клиент, оставивший отзыв: {NameOfClient}");
    }
}

public class ProductReview : Review //Отзыв Товара, простое наследование 
{
    public int ProductId { get; set; } // идентификатор продукта
    double _nowRating = new double();

    public ProductReview(int _ReviewId, string _Text, double _Rating, string _Author, DateTime _DateCreated, string _ProductName, int _ProductId, string _Tags)
        : base(_ReviewId, _Text, _Rating, _Author, _DateCreated, _ProductName, _Tags)
    {
        ProductId = _ProductId;
        _nowRating = _Rating;
        //DisplayMethodCount(this);
    }

    public void RateProduct() // переопределен для связывания рейтинга с конкретным продуктом
    {
        Console.WriteLine($"Продукту с ID {ProductId} присвоен рейтинг " + _nowRating);
    }

    public override string GetReviewDetails() // переопределен метод для получения деталей отзыва
    {
        return ($"Отзыв: {Text}, Тэги отзыва: {Tags}, Рейтинг: {_nowRating}, Автор: {Author}, Дата: {DateCreated.ToString("dd.MM.yyyy")}, Продукт: {ProductName}, ID: {ProductId}.");
    }
}

public class ServiceReview : Review // Отзыв Услуги
{
    public string _startTime { get; set; }
    public string StartTime
    {
        get { return _startTime; }
        set { _startTime = value; }
    }

    public ServiceReview(int _ReviewId, string _Text, double _Rating, string _Author, DateTime _DateCreated, string _ProductName, string _StartTime, string _Tags)
        : base(_ReviewId, _Text, _Rating, _Author, _DateCreated, _ProductName, _Tags)
    {
        StartTime = _StartTime;
        //DisplayMethodCount(this);
    }

    public override string GetReviewDetails()//переопределен для отображения времени начала услуги вместе с другими деталями отзыва. 
    {
        return ($"Услуга '{ProductName}', предоставленная в {StartTime} клиенту с именем {Author}, получила отзыв: {Text}");
    }
}

ServiceCare serviceCare = new ServiceCare(1, "Очень приятный персонал", 5.0, "Иван Иванов", "20.10.2024", "Тайский массаж", "09 сентября", "СПА-салон ЛОТОС", "Тайский массаж от мастера-профи в премиум-зале", "#массаж");
//Console.WriteLine(serviceCare.GetReviewDetail());
//Console.WriteLine(serviceCare.GetServiceDetails());
//Console.WriteLine();

ServiceRatingCare serviceRatingCare = new ServiceRatingCare(2, "Обслужили очень быстро и качественно", 5.0, "petr petrov", "22.10.2024", "Ремонт телефона", "10 сентября", "Салон техники 72ПРО", "Замена дисплея и установка защитного стекла", "Петр Петров", "#техника");
//Console.WriteLine(serviceRatingCare.GetReviewDetail());
//serviceRatingCare.DisplayClientName();
//Console.WriteLine();

ProductReview productReview = new ProductReview(3, "Товар быстро сломался", 3.2, "Семён Сидоров", DateTime.Now, "Телефон REDME 4X", 3507, "#техника");
//Console.WriteLine(productReview.GetReviewDetails());
//productReview.RateProduct();
//Console.WriteLine();

ServiceReview serviceReview = new ServiceReview(4, "Услуга оказана на хорошем уровне", 4.5, "Андрей Андреев", DateTime.Now, "Технический осмотр автомобиля", "10:00", "#авто");
//Console.WriteLine(serviceReview.GetReviewDetails());
//Console.WriteLine();


// ДОБАВЛЕНИЕ КОЛЛЕКЦИИ HASHSET
HashSet<Review> reviewSet = new HashSet<Review>();

// Добавляем объекты в HashSet
var review1 = serviceCare;
var review2 = serviceRatingCare;
var review3 = productReview;
var review4 = serviceReview;

reviewSet.Add(review1);
reviewSet.Add(review2);
reviewSet.Add(review3);
reviewSet.Add(review4);

// Попробуем добавить повторяющийся объект
var duplicateReview = new ProductReview(3, "Товар быстро сломался", 3.2, "Семён Сидоров", DateTime.Now, "Телефон REDME 4X", 3507, "#техника");
bool isAdded = reviewSet.Add(duplicateReview);
Console.WriteLine($"Возможно, был добавлен такой же отзыв? {isAdded}"); // false

// Перебираем и выводим информацию обо всех уникальных отзывах
foreach (var review in reviewSet)
{
    review.DisplayInfo();
    Console.WriteLine();
}

public class ReviewManager<T> where T : Review
{
    private List<T> _reviews = new List<T>();

    public void AddReview(T review)
    {
        _reviews.Add(review);
    }

    public void DisplayReviews()
    {
        foreach (var r in _reviews)
        {
            Console.WriteLine(r.DisplayReview());
        }
    }
}
// Пример использования
//ReviewManager<ProductReview> productReviewManager = new ReviewManager<ProductReview>();
//productReviewManager.AddReview(new ProductReview(1, "Отличный продукт!", 4.8, "Иван Иванов", DateTime.Now, "Смартфон XYZ", 12345, "#телефоны"));
//productReviewManager.AddReview(new ProductReview(2, "Разочарование в покупке", 2.5, "Анна Петрова", DateTime.Now, "Ноутбук ABC", 54321, "#ноутбуки"));
//productReviewManager.DisplayReviews();


//ДОБАВЛЕНИЕ ДЕЛЕГАТА
// Присваиваем делегату метод DisplayInfo
DisplayDelegate displayProductReview = productReview.DisplayInfo;
DisplayDelegate displayServiceReview = serviceReview.DisplayInfo;

// Вызов методов через делегаты
displayProductReview();
displayServiceReview();
Console.WriteLine();


//ДОБАВЛЕНИЕ ЛЯМБДА-ВЫРАЖЕНИЯ
List<Review> tags = new List<Review>();
var tag1 = serviceCare;
var tag2 = serviceRatingCare;
var tag3 = productReview;
var tag4 = serviceReview;

tags.Add(tag1);
tags.Add(tag2);
tags.Add(tag3);
tags.Add(tag4);

// Используем лямбда-выражение для фильтрации отзывов
var reviewTags = tags.OfType<Review>().Where(c => c.Tags == "#техника");

foreach (var tag in reviewTags)
{
    tag.DisplayTag();
}
```
